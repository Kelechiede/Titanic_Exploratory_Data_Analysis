### Dataset Overview

The Titanic dataset is a well-known dataset that provides detailed information about the passengers who were on board the RMS Titanic. The dataset consists of 12 features and 890 observations containing various attributes related to the passengers, including whether they survived the disaster or not. Below is a professional description based on the first and last few rows of the dataset:

- **PassengerId**: A unique identifier for each passenger.
- **Survived**: Indicates whether the passenger survived (1) or not (0).
- **Pclass**: The passenger class, which is an ordinal variable with values 1 (first class), 2 (second class), and 3 (third class).
- **Name**: The full name of the passenger.
- **Sex**: The gender of the passenger (male or female).
- **Age**: The age of the passenger in years. Some entries are missing.
- **SibSp**: The number of siblings or spouses the passenger had aboard the Titanic.
- **Parch**: The number of parents or children the passenger had aboard the Titanic.
- **Ticket**: The ticket number of the passenger.
- **Fare**: The fare paid by the passenger for the ticket.
- **Cabin**: The cabin number assigned to the passenger. Many entries are missing.
- **Embarked**: The port of embarkation, which can be C (Cherbourg), Q (Queenstown), or S (Southampton).

# Step 3: Data Cleaning and Preprocessing
Add a code cell to handle missing values and preprocess the data.

```python
# Check for missing values
train_df.isnull().sum()

# Handle missing values
# Fill missing Age values with the median
train_df['Age'].fillna(train_df['Age'].median(), inplace=True)

# Fill missing Embarked values with the mode
train_df['Embarked'].fillna(train_df['Embarked'].mode()[0], inplace=True)

# Drop the Cabin column due to many missing values
train_df.drop(columns=['Cabin'], inplace=True)

# Verify no missing values remain
train_df.isnull().sum()
```

# Step 4: Exploratory Data Analysis
Add code cells to generate summary statistics and visualizations.

```python
!pip install seaborn
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

# Set the style for the plots
sns.set(style='whitegrid')

# Plot survival count
plt.figure(figsize=(8, 6))
sns.countplot(data=train_df, x='Survived')
plt.title('Survival Count')
plt.show()

# Print survival count in numerical format
survival_counts = train_df['Survived'].value_counts()
print("Survival Count:")
print(survival_counts)
print(survival_counts / len(train_df) * 100, "\n")

# Plot survival rate by passenger class
plt.figure(figsize=(8, 6))
sns.countplot(data=train_df, x='Pclass', hue='Survived')
plt.title('Survival Rate by Passenger Class')
plt.show()

# Print survival rate by passenger class
pclass_survival = train_df.groupby('Pclass')['Survived'].value_counts(normalize=True).unstack()
print("Survival Rate by Passenger Class:")
print(pclass_survival)
print(pclass_survival * 100, "\n")

# Plot survival rate by sex
plt.figure(figsize=(8, 6))
sns.countplot(data=train_df, x='Sex', hue='Survived')
plt.title('Survival Rate by Sex')
plt.show()

# Print survival rate by sex
sex_survival = train_df.groupby('Sex')['Survived'].value_counts(normalize=True).unstack()
print("Survival Rate by Sex:")
print(sex_survival)
print(sex_survival * 100, "\n")

# Age distribution
plt.figure(figsize=(8, 6))
sns.histplot(data=train_df, x='Age', kde=True, hue='Survived', multiple='stack')
plt.title('Age Distribution by Survival Status')
plt.show()

# Print summary statistics for age by survival status
age_survival_summary = train_df.groupby('Survived')['Age'].describe()
print("Age Distribution by Survival Status:")
print(age_survival_summary, "\n")

# Fare distribution
plt.figure(figsize=(8, 6))
sns.histplot(data=train_df, x='Fare', kde=True, hue='Survived', multiple='stack')
plt.title('Fare Distribution by Survival Status')
plt.show()

# Print summary statistics for fare by survival status
fare_survival_summary = train_df.groupby('Survived')['Fare'].describe()
print("Fare Distribution by Survival Status:")
print(fare_survival_summary, "\n")

# Pairplot for numerical variables
sns.pairplot(train_df[['Survived', 'Pclass', 'Age', 'SibSp', 'Parch', 'Fare']], hue='Survived')
plt.show()

# Numerical summaries for pairplot variables
numerical_summaries = train_df[['Pclass', 'Age', 'SibSp', 'Parch', 'Fare']].describe()
print("Summary Statistics for Numerical Variables:")
print(numerical_summaries, "\n")
```

### Insights and Conclusions from the Exploratory Data Analysis (EDA)

#### Survival Count
The dataset reveals the following survival statistics:
- **Survived (1)**: 342 passengers (38.38%)
- **Did Not Survive (0)**: 549 passengers (61.62%)

This indicates that a significant majority of the passengers did not survive the disaster.

#### Survival Rate by Passenger Class
The survival rates by passenger class are as follows:
- **1st Class**: 62.96% survived, 37.04% did not survive
- **2nd Class**: 47.28% survived, 52.72% did not survive
- **3rd Class**: 24.24% survived, 75.76% did not survive

Passengers in the 1st class had a significantly higher survival rate compared to those in the 2nd and 3rd classes. This suggests that socio-economic status played a crucial role in survival chances.

#### Survival Rate by Sex
The survival rates by sex are as follows:
- **Female**: 74.20% survived, 25.80% did not survive
- **Male**: 18.89% survived, 81.11% did not survive

Females had a substantially higher survival rate compared to males, indicating that gender was a significant factor in determining survival, likely due to the "women and children first" evacuation policy.

#### Age Distribution by Survival Status
The summary statistics for age by survival status show:
- **Did Not Survive (0)**:
  - Mean Age: 30.03 years
  - Standard Deviation: 12.50 years
  - Median Age: 28.0 years
- **Survived (1)**:
  - Mean Age: 28.29 years
  - Standard Deviation: 13.76 years
  - Median Age: 28.0 years

The age distributions of survivors and non-survivors are quite similar, with survivors being slightly younger on average. However, both groups had the same median age, indicating that age alone was not a strong predictor of survival.

#### Fare Distribution by Survival Status
The summary statistics for fare by survival status are:
- **Did Not Survive (0)**:
  - Mean Fare: $22.12
  - Standard Deviation: $31.39
  - Median Fare: $10.50
- **Survived (1)**:
  - Mean Fare: $48.40
  - Standard Deviation: $66.60
  - Median Fare: $26.00

Survivors tended to have paid higher fares, suggesting that passengers who paid more (often those in higher classes) had better chances of survival.

#### Summary Statistics for Numerical Variables
- **Pclass**: Ranges from 1 (first class) to 3 (third class), with a mean of 2.31, indicating more passengers were in lower classes.
- **Age**: The ages range from 0.42 to 80 years, with a mean age of 29.36 years. The distribution shows a young passenger population.
- **SibSp**: Number of siblings/spouses aboard ranges from 0 to 8, with most passengers having 0 siblings/spouses aboard.
- **Parch**: Number of parents/children aboard ranges from 0 to 6, with most passengers having 0 parents/children aboard.
- **Fare**: The fares paid range from $0 to $512.33, with a mean fare of $32.20. This wide range highlights the diversity in ticket prices.

### Conclusions
The exploratory data analysis of the Titanic dataset provides several key insights:
- **Survival Rates**: The majority of passengers did not survive. However, survival rates varied significantly across different classes and genders. First-class passengers and females had notably higher survival rates.
- **Economic and Social Factors**: Socio-economic status, as indicated by passenger class and fare paid, played a critical role in survival. Higher-class passengers, who paid higher fares, had better survival chances.
- **Demographics**: While age did not appear to be a major determinant of survival, the data shows a slightly younger average age for survivors. The median age was the same for both survivors and non-survivors.
- **Family Relationships**: The number of family members aboard (siblings/spouses and parents/children) varied, but most passengers traveled alone or with one family member.

These insights suggest that in the tragic event of the Titanic, socio-economic status and gender significantly influenced the likelihood of survival. This analysis can guide further predictive modeling and deeper exploration into the factors affecting survival outcomes on the Titanic.
